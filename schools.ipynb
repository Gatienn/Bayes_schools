# Sampler de Gibbs

```python
def mu_ij(alphaj, beta, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool) : 
    mu = alphaj[0] + alphaj[1]*LRT + alphaj[2]*VR1+ beta[0]*LRT**2 + beta[1]*VR2 + \
    beta[2]*Girl + beta[3]*GirlSchool + beta[4]*BoySchool + beta[5]*CESchool + beta[6]*RCSchool + beta[7]*OtherSchool
    return(mu)

def tau_ij(theta, phi, LRT) :
    return(np.exp(theta + phi*LRT))
```

```python
def logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school,
              theta, phi, beta, M, T, alpha) :
    n = len(Y)
    like = 0
    for i in range(n) : 
        j = int(school[i]) - 1
        mu = mu_ij(alpha[:,j], beta, LRT[i], VR1[i], VR2[i], Girl[i], GirlSchool[i], BoySchool[i], CESchool[i], RCSchool[i], OtherSchool[i])
        tau = tau_ij(theta, phi, LRT[i])
        like +=  -0.5*tau*(Y[i]-mu)**2
    return(like)
```

```python
def sampler(nchain, data, init, init_alpha, init_M, init_T, prop_sd) :
    
    # initialisation of the chains
    chain = np.zeros(shape = (nchain+1, 10)) # theta, phi, beta*8 
    chain_alpha = np.zeros(shape = (nchain+1, 3, 38)) # alpha1, 2, 3 * m=38
    chain_M = np.zeros(shape = (nchain+1, 3)) # for M
    chain_T = np.zeros(shape = (nchain+1, 3, 3)) # for T
    chain[0,:] = init
    chain_alpha[0,:,:] = init_alpha
    chain_M[0,:] = init_M
    chain_T[0, :,:] = init_T
    
    # acceptance rate : M&T have no acceptance rate since we have a conjuguate posterior distribution
    acceptance_rate = np.zeros(shape = 10)
    acceptance_rate_alpha = np.zeros(shape = 38) 
    
    # hyper-parameters 
    m = 38
    n = 1978
    sigma_0 = 1/0.0001 # given in volume
    T_1 = np.ones((3,3))/sigma_0 # we decided it 
    R = np.array([[0.1, 0.005, 0.005], [0.005, 0.01, 0.005], [0.005, 0.005, 0.01]]) # given in volume 
    d=3
    
    # data 
    Y = data[:,0]
    LRT = data[:,1]
    school = data[:,2]
    Girl = data[:,3]
    GirlSchool = data[:,4]
    BoySchool = data[:,5]
    CESchool = data[:,6]
    RCSchool = data[:,7]
    OtherSchool = data[:,8]
    VR1 = data[:,9]
    VR2 = data[:,10]
    
    print("commenc")
    
    for k in range(nchain):
        
        ## for alphas :
        for j in range(m) :
            alpha_old = chain_alpha[k,:,:]
            alpha_j_old = chain_alpha[k,:,j]
            prop_j = np.random.multivariate_normal(alpha_j_old, np.diag([1,1,1])*prop_sd[j])
            prop = np.copy(alpha_old)
            prop[:,j] = prop_j 
            
            C = np.linalg.inv(chain_T[k,:,:])
            
            num = stats.multivariate_normal.pdf(prop_j, mean=chain_M[k,:], cov=C)+ logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k,0], chain[k,1], chain[k,2:10], chain_M[k,:], chain_T[k,:,:], prop)
            den = stats.multivariate_normal.pdf(alpha_j_old , mean=chain_M[k,:], cov=C)+ logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k,0], chain[k,1], chain[k,2:10], chain_M[k,:], chain_T[k,:,:], alpha_old)
            
            prob_acceptation = np.exp(num - den)
            if (prob_acceptation > np.random.uniform()).all():
                chain_alpha[k+1,:,j] = prop_j
                acceptance_rate_alpha[j] += 1
            else :
                chain_alpha[k+1,:,j] = alpha_j_old
        
        
        # update theta
        theta_old = chain[k,0]
        prop = np.random.normal(theta_old, prop_sd[m])
        
        num = - prop**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, prop, chain[k,1], chain[k,2:10], chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
        den = - theta_old**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, theta_old, chain[k,1], chain[k,2:10], chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
        
        prob_acceptation = np.exp(num - den)
        if prob_acceptation > np.random.uniform():
            chain[k+1,0] = prop
            acceptance_rate[0] +=1
        else :
            chain[k+1,0] = theta_old
        
        
        # update phi
        phi_old = chain[k,1]
        prop = np.random.normal(phi_old, prop_sd[m+1])
        
        num = - prop**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k+1,0], prop, chain[k,2:10], chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
        den = - theta_old**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k+1,0], phi_old, chain[k,2:10], chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
        
        prob_acceptation = np.exp(num - den)
        if prob_acceptation > np.random.uniform():
            chain[k+1,1] = prop
            acceptance_rate[1] +=1
        else :
            chain[k+1,1] = phi_old
        
        
        # update beta
        for l in range(8) :
            beta_old = chain[k, 2:10]
            beta_l_old = chain[k,2+l]
            prop_l = np.random.normal(beta_l_old, prop_sd[m+2+l])
            prop = np.copy(beta_old)
            prop[l] = prop_l
        
            num = - prop_l**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k+1,0], chain[k+1,1], prop, chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
            den = - beta_l_old**2/(2*sigma_0**2) + logLike_y(Y, LRT, VR1, VR2, Girl, GirlSchool, BoySchool, CESchool, RCSchool, OtherSchool, school, chain[k+1,0], chain[k+1,1], beta_old , chain_M[k,:], chain_T[k,:,:], chain_alpha[k+1,:,:])
        
            prob_acceptation = np.exp(num - den)
            if prob_acceptation > np.random.uniform():
                chain[k+1,2+l] = prop_l
                acceptance_rate[2+l] +=1
            else :
                chain[k+1,2+l] = beta_l_old
            
            
        # update M
        mu_tilde = np.linalg.inv(T_1 + m*chain_T[k,:,:])@chain_T[k,:,:]@np.sum(chain_alpha[k+1,:,:], axis=1)
        T_tilde = T_1 + m*chain_T[k,:,:]
        cov_tilde = np.linalg.inv(T_tilde)
        chain_M[k+1,:] = np.random.multivariate_normal(mu_tilde, cov_tilde)
        
        
        # update T
        mat = np.zeros((3,3))
        M = chain_M[k+1,:]
        for j in range(m) :
            aj = chain_alpha[k+1,:,j]
            mat += (aj-M)@np.transpose(aj-M)
        V_tilde = np.linalg.inv(np.linalg.inv(R) + mat)
        nu_tilde = m+d
        chain_T[k+1,:,:] = stats.wishart.rvs(nu_tilde, scale = V_tilde)

        print(k)
        
    print("fini")
    return(chain, chain_alpha, chain_M, chain_T, 1/nchain*acceptance_rate, 1/nchain*acceptance_rate_alpha)
    
    
```
